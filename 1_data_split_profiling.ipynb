```python
import pandas as pd
from ydata_profiling import ProfileReport
```

```python
# GLOBAL VAR
PATH = './'
TRAIN_FRAC = 0.80
RND_STATE = 123
```

```python
# READ DATA
dataset = pd.read_csv(PATH+'all_data.csv')
dataset.head()
```

```python
# SPLIT DATA
df = dataset.sample(frac=TRAIN_FRAC, random_state=RND_STATE)
df_unseen = dataset.drop(df.index)

df.reset_index(inplace=True, drop=True)
df_unseen.reset_index(inplace=True, drop=True)

print('Data for Modeling: ' + str(df.shape))
print('Unseen Data For Predictions ' + str(df_unseen.shape))
```

```python
# SAVE DATASETS
df.to_csv(PATH+'data_training.csv', index=False)
df_unseen.to_csv(PATH+'data_testing.csv', index=False)
```

```python
# PROFILE TRAINING DATASET
profile = ProfileReport(df, title="Training Data Exploration Report", explorative=True)
profile.to_file(PATH+"training_data_exploration_report.html")
```
